```python
from pathlib import Path
import json
import pickle
import numpy as np
import pandas as pd
from bertopic import BERTopic
```

```python
base_dir = Path("./")
```

```python
def load_results(output_dir):
    with open(output_dir / "params.json", 'r') as f:
        params = json.load(f)

    topic_model = BERTopic.load(output_dir / "saved_model", embedding_model=params["embedding_model_name"])

    with open(output_dir / "docs.pkl", 'rb') as f:
        docs = pickle.load(f)

    embeddings = np.load(output_dir / f"embeddings_{params['embedding_model_name']}.npy")

    chunks_with_topic = pd.read_csv(output_dir/"chunks_with_topic.csv")

    return topic_model, embeddings, docs, chunks_with_topic
```

# La_Voyeuse_interdite

```python
data_root = Path(base_dir) / "topic_modeling_data"
experiment_name = "nina_Bouraoui_La_Voyeuse_interdite"
output_dir = Path(base_dir) / f"topic_modeling_output/{experiment_name}"
```

```python
topic_model, embeddings, docs, chunks_with_topic = load_results(output_dir)
```

```python
topic_model.get_topic_info()
```

```python
topic_model.visualize_documents(docs, topics=[1, 0, 11, 14, 15, 27])
```
